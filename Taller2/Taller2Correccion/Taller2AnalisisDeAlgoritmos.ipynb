```python
class Operaciones:
    @staticmethod
    def factorial(n):
        if n == 0 or n == 1:  # O(1)
            return 1
        else:
            resultado = 1  # O(1)
            for i in range(2, n + 1):  # O(n)
                resultado *= i  # O(1)
            return resultado  # O(1)

    @staticmethod
    def multiplicar_numeros(lista):
        resultado = 1  # O(1)
        for numero in lista:  # O(n)
            resultado *= numero  # O(1)
        return resultado  # O(1)


class PrinDeCont:
    @staticmethod
    def permutaciones_repetidas(n):
      p_valor = 0  # O(1)
      list_num = []  # O(1)

      while True:  # O(n)
          pregunta_repite = input("Cuantos elementos se repiten: ")  # O(1)
          if pregunta_repite.isdigit() and pregunta_repite != "0":  # O(1)
              p_valor = int(pregunta_repite)  # O(1)
              break  # O(1)
          else:
              print("Lo que ingresaste debe ser un numero entero y diferente de 0, vuelve a intentarlo.\n")  # O(1)

      for i in range(0, p_valor):  # O(p_valor)
          while True:  # O(p_valor)
              num_valor = 0  # O(1)
              num = input("Ingresa el valor repetido: ")  # O(1)
              if num.isdigit() and num != "0":  # O(1)
                  num_valor = int(num)  # O(1)
                  break  # O(1)
              else:
                  print("Lo que ingresaste debe ser un numero entero y diferente de 0, vuelve a intentarlo.\n")  # O(1)

          num_factorial = Operaciones.factorial(num_valor)  # O(num_valor)
          list_num.append(num_factorial)  # O(1)
      return Operaciones.factorial(n) // Operaciones.multiplicar_numeros(list_num)  # O(n)

      @staticmethod
      def variaciones(n, r):
          print("\nPara tu ejercicio necesitas una variacion")  # O(1)
          pregunta = input("Se repite? (s/n): ")  # O(1)
          if pregunta.lower() == "s":
              return n**r  # O(1)
          elif pregunta.lower() == "n":
              return Operaciones.factorial(n) // Operaciones.factorial(n-r)  # O(n)
          else:
              print("\nOpcion no valida\n")  # O(1)
              return PrinDeCont.variaciones(n, r)  # O(n)


class Main:
    @staticmethod
    def main():
      while True:  # O(n)
          n_valor = obtener_entero("Por favor, ingresa el valor de 'n': ")  # O(1)
          r_valor = obtener_entero("Por favor, ingresa el valor de 'r': ")  # O(1)

          if n_valor >= r_valor:  # O(1)
              pregunta_prinmulti = input("Escoja 1 para solucionarlo solo con principio multiplicativo o 2 para que sea con otro principio de conteo: ")  # O(1)
              if pregunta_prinmulti == "2":  # O(1)
                  respuesta_orden = input("¿El orden importa para tu ejercicio? (s/n): ")  # O(1)
                  respuesta_elementos = input("¿Todos los elementos son considerados para tu ejercicio? (s/n): ")  # O(1)

                  if respuesta_orden.lower() == "s" and respuesta_elementos.lower() == "s":  # O(1)
                      print("\nPara tu ejercicio necesitas una permutacion")  # O(1)
                      pregunta = input("Se repite (s/n): ")  # O(1)
                      if pregunta.lower() == "s":  # O(1)
                          print(f"\nEl resultado de la permutacion es: {PrinDeCont.permutaciones_repetidas(n_valor)}\n")  # O(1)
                      elif pregunta.lower() == "n":  # O(1)
                          print(f"\nEl resultado de la permutacion es: {PrinDeCont.permutaciones_no_repetidas(n_valor)}\n")  # O(1)
                      else:
                          print("\nOpcion no valida\n")  # O(1)

                  elif respuesta_orden.lower() == "s" and respuesta_elementos.lower() == "n":  # O(1)
                      print(f"\nEl resultado de las variaciones es: {PrinDeCont.variaciones(n_valor, r_valor)}\n")  # O(1)
                  elif respuesta_orden.lower() == "n" and respuesta_elementos.lower() == "n":  # O(1)
                      print(f"\nEl resultado de las combinaciones es: {PrinDeCont.combinaciones(n_valor, r_valor)}\n")  # O(1)
                  elif respuesta_orden.lower() == "n" and respuesta_elementos.lower() == "s":  # O(n)
                      print("\nLo siento, no hay principio de conteo para tus respuestas")  # O(1)
                  else:
                      print("\nLo que ingresaste no es válido. Por favor, asegúrate de responder 's' o 'n' a las preguntas.\n")  # O(1)
              elif pregunta_prinmulti == "1":  # O(1)
                  print(f"\nEl resultado del principio multiplicativo es es: {PrinDeCont.prinmulti(n_valor, r_valor)}\n")  # O(1)
              else:
                  print("Opcion no valida")  # O(1)
          else:
              print("No se puede ejecutar, n debe ser mayor o igual a r")  # O(1)

          Resp = input("¿Deseas realizar otra operación? (s/n): ")  # O(1)
          if Resp.lower() == "s":  # O(1)
              print("\n¡Entendido, continuemos!\n")  # O(1)
          else:
              print("\nGracias por usar el Programa de Principios de Conteo de Codex. ¡Hasta luego!\n")  # O(1)
              break  # O(1)


def obtener_entero(mensaje):
    while True:  # O(n)
        entrada = input(mensaje)  # O(1)
        if entrada.isdigit() and entrada != "0":  # O(1)
            return int(entrada)  # O(1)
        else:
            print("Lo que ingresaste debe ser un numero entero y diferente de 0, vuelve a intentarlo.\n")  # O(1)



mensaje = """
---------------------------------------------------
|              ¡Bienvenido al Programa de          |
|          Principios de Conteo de Codex!          |
|                                                  |
| Los principios de conteo son técnicas en         |
| combinatoria para calcular el número de          |
| posibles resultados en un experimento o          |
| situación.                                       |
|                                                  |
| El principio fundamental del conteo establece    |
| que si una acción sepuede realizar de 'n'        |
| maneras diferentes y otra de 'r', entonces hay   |
| 'n * r' maneras diferentes de realizar ambas     |
| acciones.                                        |
|                                                  |
| Existen dos enfoques principales para el         |
| conteo: permutaciones y combinaciones. Las       |
| permutaciones consideran el orden en el que se   |
| eligen los elementos, mientras que las           |
| combinaciones no consideran el orden.            |
|                                                  |
---------------------------------------------------
"""
print(mensaje)

if __name__ == "__main__":
    Main.main()
```
