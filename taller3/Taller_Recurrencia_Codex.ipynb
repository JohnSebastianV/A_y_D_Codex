```python
import mpmath
from fractions import Fraction

class TerminoN:
    def __init__(self):
        self.inicio = 0
        self.n = 0
        self.k1 = 0
        self.k2 = 0
        self.r1 = 0
        self.r2 = 0

    def solicitar_inicio(self):
        while True:
            inicio_input = input("Ingrese el valor de inicio (n>=?): ")
            try:
                self.inicio = float(inicio_input)
                if self.inicio < 0:
                    raise ValueError("El valor de inicio debe ser mayor o igual a 0.")
                break
            except ValueError:
                print("Por favor, ingrese un número válido para el inicio.")

    def solicitar_termino_n(self):
        while True:
            n_input = input("Ingrese el término de an: ")
            try:
                self.n = float(n_input)
                if self.n < 0:
                    raise ValueError("El término an debe ser mayor o igual a 0.")
                break
            except ValueError:
                print("Por favor, ingrese un número válido para an.")

    def solicitar_Ks_y_Rs(self):
        while True:
            try:
                self.k1 = mpmath.mpf(input("Ingrese el término de K1: "))
                self.k2 = mpmath.mpf(input("Ingrese el término de K2: "))
                self.r1 = mpmath.mpf(input("Ingrese el término de R1: "))
                self.r2 = mpmath.mpf(input("Ingrese el término de R2: "))
                break
            except ValueError:
                print("Por favor, ingrese números válidos para K1, K2, R1 y R2.")

    def calcular_formula_diferentes(self):
        try:
            K1R1 = self.k1 * self.r1 ** self.n
            K2R1 = self.k2 * self.r2 ** self.n
            formula = K1R1 + K2R1
        except OverflowError:
            formula = None
        return formula

    def calcular_formula_iguales(self):
        try:
            K1R1 = self.k1 * self.r1 ** self.n
            K2R1 = self.k2 * self.n * self.r2 ** self.n
            formula = K1R1 + K2R1
        except OverflowError:
            formula = None
        return formula

    def mostrar_resultado(self, resultado):
        if resultado is None:
            print("El resultado es demasiado grande para ser representado.")
        else:
            print("El resultado es:", resultado)

    def main(self):
        self.solicitar_Ks_y_Rs()
        self.solicitar_inicio()
        while True:
            self.solicitar_termino_n()
            try:
                if self.n >= self.inicio:
                    if self.r1 == self.r2:
                        resultado = self.calcular_formula_iguales()
                        self.mostrar_resultado(resultado)
                    else:
                        resultado = self.calcular_formula_diferentes()
                        self.mostrar_resultado(resultado)
                else:
                    raise ValueError("Lo siento, pero an debe ser mayor o igual al inicio")
            except ValueError as e:
                print(e)
            continuar = input("¿Desea continuar con los mismos valores y solo cambiar an? (si/no): ")
            if continuar.lower() == "si":
                pass
            elif continuar.lower() == "no":
                break
            else:
                print("Opción no válida. Se solicitará nuevamente n")
        pregunta = input("¿Desea hacer otra operación? (si/no): ")
        if pregunta.lower() == "si":
            self.main()

if __name__ == "__main__":
    TerminoN().main()
```
